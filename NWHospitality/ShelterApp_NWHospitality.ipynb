```python
import pandas as pd
import numpy as np
```

## Reading Data

```python
df = pd.read_csv("Homeless Resources in Washington State - Curated by Northwest Hospitality-Grid view.csv")
```

```python
df.head()
```

```python
df_columns = df[["Summary", "Full Address", "Category"]].copy()
```

```python
df_columns.head()
```

## Cleaning

```python
df_columns["Full Address"].replace('', np.nan, inplace=True)
```

```python
df_columns.shape
```

```python
df_columns.dropna(subset=["Full Address"], inplace=True)
```

```python
df_columns.shape
```

```python
df_columns["zip"] = df["Full Address"].str[-5:]
```

```python
df_columns.head()
```

```python
df_columns["state"] = "WA"
```

```python
df_columns["address"] = df_columns["Full Address"]
```

```python
df_columns.shape
```

```python
df_columns = df_columns.replace('\n','', regex=True)
```

```python
df_columns.head()
```

```python
df_columns["city"] = df_columns["address"].apply(lambda a: a[a.find(",")+1: a.find(",")+1+a[a.find(",")+1:].find(",")])
```

```python
df_columns.head()
```

```python
df_columns["address"] = df_columns["address"].apply(lambda a: a[:a.find(",")])
```

```python
df_columns.head()
```

```python
df_columns["Category"].unique()
```

```python
df_columns.rename(columns={'Summary':'name'}, inplace=True)
```

```python
df_columns.head()
```

```python
df_columns["Summary"] = df_columns["name"]
```

```python
df_columns.head()
```

```python
df_columns["name"] = df_columns["name"].apply(lambda n : n[:n.find("-")-1])
```

```python
df_columns.head()
```

## Exporting to MongoDB

```python
from pymongo import MongoClient, TEXT 
import os
```

```python
import json
```

```python
os.environ["DBUSERNAME"] = "amazonads"
os.environ["DBPW"] = "AmazonAds2020"
```

```python
client = MongoClient(
    "mongodb+srv://" + os.environ.get('DBUSERNAME') + ":" + os.environ.get('DBPW')
    + "@shelter-rm3lc.azure.mongodb.net/shelter?retryWrites=true&w=majority"
)['shelter']
```

```python
df_rows = json.loads(df_columns.T.to_json()).values()
```

```python
db = client
db_coll = db["tmpNWHospitality"]
db_coll.insert_many(df_rows)
```
